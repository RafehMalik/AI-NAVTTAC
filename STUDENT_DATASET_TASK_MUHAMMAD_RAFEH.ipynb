```python
#  Import libraries
import pandas as pd

#Load the dataset
from google.colab import files
uploaded = files.upload()
```

```python
# Read CSV file into DataFrame
df = pd.read_csv("student.csv")
```

```python
df.head()
```

```python
# Display missing values count per column
print("Missing values in each column:\n", df.isnull().sum())
```

```python
#Convert 'Age' to numeriforcing non-numeric to NaN, and fill with mean
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')
df['Age'].fillna(df['Age'].mean(), inplace=True)
print(df['Age'].head())
```

```python
#Clean 'Score' column, convert to numeric, fill missing with median
df['Score'] = pd.to_numeric(df['Score'], errors='coerce')
df['Score'].fillna(df['Score'].median(), inplace=True)
print(df['Score'].head())
```

```python
#Drop rows where Name or City is missing
df.dropna(subset=['Name', 'City'], inplace=True)
print(df.head())
```

```python
#Remove leading/trailing whitespaces in Name and City
df['Name'] = df['Name'].str.strip()
df['City'] = df['City'].str.strip()
print(df[['Name', 'City']].head())
```

```python
#Convert Name column to title case
df['Name'] = df['Name'].str.title()
```

```python
#Convert JoinedDate to datetime format (handling invalids)
df['JoinedDate'] = pd.to_datetime(df['JoinedDate'], errors='coerce')
print(df['JoinedDate'].head())
```

```python
#Remove duplicate rows
df.drop_duplicates(inplace=True)
df.head()
```

```python
#Replace missing City with 'Unknown'
df['City'].fillna('Unknown', inplace=True)
```

```python
#Filter and list students with Score >= 90
high_scorers = df[df['Score'] >= 90]
print("\nStudents with Score >= 90:\n", high_scorers[['Name', 'Age', 'Score']])
```

```python
#Remove rows where Age is still NaN (invalid ages)
df = df[df['Age'].notnull()]
```

```python
#Rename all columns to lowercase
df.columns = df.columns.str.lower()
print(df.columns)
```

```python
df.head(10)
```

```python
#Save cleaned data to a new CSV file
df.to_csv("students_cleaned.csv", index=False)
print("\nCleaned dataset saved as 'students_cleaned.csv'.")
```
