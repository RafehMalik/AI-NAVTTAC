```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
```

```python
iris = load_iris()

df = pd.DataFrame(iris.data, columns=iris.feature_names)
X = df.drop('petal length (cm)', axis=1)  # Features (excluding the target variable)
y = df['petal length (cm)']

print(df.head())
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
coefficients = model.coef_
intercept = model.intercept_
```

```python
y_pred = model.predict(X)
```

```python
print("Coefficients:", coefficients)
print("Intercept:", intercept)
print("Predictions for the first 5 samples:", y_pred[:5])
```

```python
# Create a scatter plot of the actual data points
plt.scatter(X['sepal length (cm)'], y, color='blue', label='Data points')


plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.title('Scatter plot with Best Fit Line')
plt.legend()

plt.show()
```
